```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
mouse_drug_data_to_load = "data/mouse_drug_data.csv"

clinical_data = pd.read_csv(clinical_trial_data_to_load)
mouse_data = pd.read_csv(mouse_drug_data_to_load)

```

```python
clinical_data
```

```python
mouse_data
```

```python
combined_clinical_mouse_data = pd.merge(clinical_data, mouse_data, how ="outer", on="Mouse ID")
combined_clinical_mouse_data.head()
```

# 1. Tumor Response to Treatment

```python
drug_time_group = combined_clinical_mouse_data.groupby(['Drug','Timepoint'])
drug_time_group
```

```python
tumor_volume_mean = round(drug_time_group["Tumor Volume (mm3)"].mean(),3)
tumor_volume_mean
```

```python
tumor_response_1 = pd.DataFrame(tumor_volume_mean)
tumor_response_1 = tumor_response_1.reset_index()
tumor_response_1
```

```python
tumor_volume_sem = round(drug_time_group["Tumor Volume (mm3)"].sem(),3)
tumor_volume_sem
```

```python
tumor_response_2 = pd.DataFrame(tumor_volume_sem)
tumor_response_2 = tumor_response_2.reset_index()
tumor_response_2
```

```python
tumor_mean_pivot = tumor_response_1.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
tumor_mean_pivot
```

```python
tumor_sem_pivot = tumor_response_2.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
tumor_sem_pivot.head()
```

```python
plt.errorbar(tumor_mean_pivot.index,tumor_mean_pivot["Capomulin"], yerr=tumor_sem_pivot["Capomulin"],color='y',
             marker='o',markersize=5,linestyle="dashed",linewidth=1, label="Capomulin")

plt.errorbar(tumor_mean_pivot.index,tumor_mean_pivot["Infubinol"], yerr=tumor_sem_pivot["Infubinol"],color='m',
             marker='2',markersize=5,linestyle="dashed",linewidth=1, label="Infubinol")

plt.errorbar(tumor_mean_pivot.index,tumor_mean_pivot["Ketapril"], yerr=tumor_sem_pivot["Ketapril"],color='c',
             marker='s',markersize=5,linestyle="dashed",linewidth=1, label="Ketapril")

plt.errorbar(tumor_mean_pivot.index,tumor_mean_pivot["Placebo"], yerr=tumor_sem_pivot["Placebo"],color='k',
             marker='d',markersize=5,linestyle="dashed",linewidth=1, label="Placebo")

plt.title("Graph #1: Tumor Response to Treatment")
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.ylim(30,75)
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid(True)
plt.savefig("../Images/Tumor Response.png")
plt.show()
```

# 2. Metastatic Response to Treatment

```python
metastatic_mean = round(drug_time_group["Metastatic Sites"].mean(),3)
metastatic_mean
```

```python
metastatic_response_1 = pd.DataFrame(metastatic_mean)
metastatic_response_1 = metastatic_response_1.reset_index()
metastatic_response_1
```

```python
metastatic_sem = round(drug_time_group["Metastatic Sites"].sem(),3)
metastatic_sem
```

```python
metastatic_response_2 = pd.DataFrame(metastatic_sem)
metastatic_response_2 = metastatic_response_2.reset_index()
metastatic_response_2
```

```python
metastatic_mean_pivot = metastatic_response_1.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
metastatic_mean_pivot.head()
```

```python
metastatic_sem_pivot = metastatic_response_2.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
metastatic_sem_pivot.head()
```

```python
plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Capomulin"], yerr=metastatic_sem_pivot["Capomulin"],color='y',
             marker='o',markersize=5,linestyle="dashed",linewidth=1, label="Capomulin")

plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Infubinol"], yerr=metastatic_sem_pivot["Infubinol"],color='m',
             marker='2',markersize=5,linestyle="dashed",linewidth=1, label="Infubinol")

plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Ketapril"], yerr=metastatic_sem_pivot["Ketapril"],color='c',
             marker='s',markersize=5,linestyle="dashed",linewidth=1, label="Ketapril")

plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Placebo"], yerr=metastatic_sem_pivot["Placebo"],color='k',
             marker='d',markersize=5,linestyle="dashed",linewidth=1, label="Placebo")


plt.title("Graph #2: Metastatic Spread During Treatment")
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Metastatic Sites')
plt.ylim(0.0,4.0)
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid(True)
plt.savefig("../Images/Metastatic Spread.png")
plt.show()
```

# 3. Survival Rates

```python
mouse_count = round(drug_time_group["Mouse ID"].count(),3)
mouse_count
```

```python
mouse_response_1 = pd.DataFrame(mouse_count)
mouse_response_1 = mouse_response_1.reset_index()
mouse_response_1
```

```python
metastatic_count_pivot = mouse_response_1.pivot(index="Timepoint",columns="Drug",values="Mouse ID")
metastatic_count_pivot
```

```python
plt.plot(metastatic_count_pivot["Capomulin"]/25*100, color='y',
             marker='o',markersize=5,linestyle="dashed",linewidth=1, label="Capomulin")

plt.plot(metastatic_count_pivot["Infubinol"]/25*100,color='m',
         marker='2',markersize=5,linestyle="dashed",linewidth=1, label="Infubinol")

plt.plot(metastatic_count_pivot["Ketapril"]/25*100,color='c',
         marker='s',markersize=5,linestyle="dashed",linewidth=1, label="Ketapril")

plt.plot(metastatic_count_pivot["Placebo"]/25*100,color='k',
         marker='d',markersize=5,linestyle="dashed",linewidth=1, label="Placebo")

plt.title("Graph #3: Survival During Treatment")
plt.xlabel('Time(Days)')
plt.ylabel('Survival Rate(%)')
plt.ylim(20,105)
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid(True)
plt.savefig("../Images/Survival Rate.png")
plt.show()
```

# 4. Summary Bar Graph

```python
percent_change = (tumor_mean_pivot.iloc[-1]-tumor_mean_pivot.iloc[0])/tumor_mean_pivot.iloc[0]*100
percent_change
```

```python
percent_ch = (percent_change["Capomulin"], percent_change["Infubinol"],
              percent_change["Ketapril"],percent_change["Placebo"],) 
percent_ch
```

```python
drug_name = ["Capomulin","Infubinol", "Ketapril","Placebo"]
```

```python
plt.figure(figsize=(20,15))
width = 0.35
x_axis = np.arange(len(percent_ch))

fig, ax = plt.subplots()

rects1 = ax.bar(drug_name[0],percent_ch[0],color="g")
rects2 = ax.bar(drug_name[1],percent_ch[1],color="r")
rects3 = ax.bar(drug_name[2],percent_ch[2],color="r")
rects4 = ax.bar(drug_name[3],percent_ch[3],color="r")


ax.set_title('Graph #4: Tumor Change over 45 Day Treatment')
ax.set_ylabel('% Tumor Volume Change')
ax.set_xticks(x_axis + width / 2)
ax.set_xticklabels(('Capomulin','Infubinol','Ketapril','Placebo'))


def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # If we can fit the label above the column, do that;
        # otherwise, put it inside the column.
        if p_height > 0.20: # arbitrary; 95% looked good to me.
            label_position = height - (y_height * 0.05)
        else:
            label_position = height + (y_height * 0.01)

        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',
                ha='center', va='bottom', color='white')

autolabel(rects1, ax)
autolabel(rects2, ax)
autolabel(rects3, ax)
autolabel(rects4, ax)


plt.tight_layout()
plt.savefig("../Images/Tumor Change.png")
plt.show()
```

# Observations:

1. From the Tumor Response Treatment plot (graph #1), it can be inferred that when mice are treated with the drug "Capomulin", the tumor volume significatly decreases over time. The opposite results can be observed when mice are treated with "Infubinol" and "Ketapril" drugs. Compared to the Placebo treamtment, "Infubinol" leads to a slightly smaller tumor volume. On the other hand, it seems that when mice are treated with "Ketapril", the tumor volume slightly increases more than the Placebo. 

2. From graph #2 it can be observed that there are fewer metastatic sites for mice treated with "Capomulin"than those mice treated with "Infubinol" and "Ketapril". For all three drugs, the number of metastatic sites increased over time but in the case of "Ketapril" after 45 days of treatment, it seems that the mice had more metastatic sites than if they weren't been treated with any drug (as the Placebo results indicate).

3. The survival rate is consistent for mice treated with "Capomulin". For mice treated with "Infubinol" the survival rate is noticably lower than the other drugs. (Graph #3)

4. Finally, it can observed that if mice are treated with "Capomulin" for the 45 day treatment period, their tumor volume decreased on average 19% (graph #1, #4).  This is a much more favorable result than for "Infuinol" and "Ketapril" which showed the tumor volume increasing over that period (for  "Infubinol" (46%) and "Ketapril" (57%)).  Indeed, the treatment with "Ketapril" appears worse than just taking a Placebo.   

5. It can be concluded that the treatment with "Capomulin" is more beneficial to treating cancer in mice beacause there is a reduction in the size of the tumors and less number of metastatic sites than the corresponding treatments with "Infubinol" and "Ketapril". Consequently, "Campomulin" has a higher and more consistent survival rate than the other treatments.  The worst drug of the those studied is "Ketapril" because is caused the tumors to increase in size even greater than just taking a placebo, it resulted in the most metastatic sites and it had a low survival rate.
